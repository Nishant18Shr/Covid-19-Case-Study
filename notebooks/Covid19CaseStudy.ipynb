Step 2: Handle Missing Values (Forward Fill)

```python
def clean_dataset(df):

    df = df.drop(columns=["Province/State", "Lat", "Long"])

    df = df.groupby("Country/Region").sum()

    df = df.T

    df.index = pd.to_datetime(df.index, format='%m/%d/%y')

    df = df.sort_index()

    df = df.ffill()

    # Clean labels
    df.columns.name = None
    df.index.name = None

    return df
```

```python
confirmed_clean = clean_dataset(covid_19_confirmed)
deaths_clean = clean_dataset(covid_19_deaths)
recovered_clean = clean_dataset(covid_19_recovered)
```

```python
def fill_province(df):
    df["Province/State"] = df["Province/State"].fillna("All Provinces")
    df["Province/State"] = df["Province/State"].replace("", "All Provinces")
    return df

confirmed_df = fill_province(covid_19_confirmed)
deaths_df = fill_province(covid_19_deaths)
recovered_df = fill_province(covid_19_recovered)
```

# 5.1 Peak Daily New Cases (Germany, France, Italy)

Daily new cases were computed using first-order difference. Italy/France/Germany were compared to identify the peak surge.

```python
# Use already cleaned confirmed dataset
df = confirmed_clean.copy()

# Select countries
countries = ["Germany", "France", "Italy"]
df_sel = df[countries]

# Calculate daily new cases
daily_cases = df_sel.diff()

# Find peak values
peak_info = {}

for country in countries:
    peak_value = daily_cases[country].max()
    peak_date = daily_cases[country].idxmax()

    peak_info[country] = (peak_value, peak_date)

# Print results
for country, (value, date) in peak_info.items():
    print(f"{country}: Peak = {value} on {date.date()}")
```

# 5.2 Recovery Rate (Canada vs Australia)

Recovery rate was calculated as recovered/confirmed on Dec 31, 2020. Country with higher rate showed better recovery management.

```python
# Use cleaned datasets
confirmed = confirmed_clean
recovered = recovered_clean

# Date of interest
date = "2020-12-31"

# Convert to datetime index
date = pd.to_datetime(date)

# Get values
countries = ["Canada", "Australia"]

results = {}

for country in countries:
    rec = recovered.loc[date, country]
    conf = confirmed.loc[date, country]

    rate = rec / conf if conf != 0 else 0
    results[country] = rate

# Print
for c, r in results.items():
    print(f"{c}: Recovery Rate = {r:.2%}")
```

# 5.3 Death Rate Distribution (Canada Provinces)

Province-level death rates were computed for Canada using latest data. Highest and lowest provinces were identified.

```python
# Use ORIGINAL datasets
conf = confirmed_df.copy()
death = deaths_df.copy()

# Filter Canada
conf_can = conf[conf["Country/Region"] == "Canada"]
death_can = death[death["Country/Region"] == "Canada"]

# Latest date column
latest_date = conf_can.columns[-1]

# Set Province as index
conf_can = conf_can.set_index("Province/State")
death_can = death_can.set_index("Province/State")

# Calculate death rate
death_rate = death_can[latest_date] / conf_can[latest_date]

# Remove NaN or inf
death_rate = death_rate.replace([float('inf'), -float('inf')], 0).fillna(0)

# Find highest and lowest
highest = death_rate.idxmax(), death_rate.max()
lowest = death_rate.idxmin(), death_rate.min()

print("Highest:", highest)
print("Lowest:", lowest)
```

# 6.1 Transform deaths dataset (Wide -> Long -> Clean Time-Series)

The dataset was transformed from wide to long format using melt(), followed by aggregation and pivoting to obtain a time-series format with dates as rows and countries as columns.


* **Data transformation from wide to long format improves flexibility for analysis and enables efficient aggregation and time-series operations.**

```python
# Start from original deaths_df (not cleaned one)
df = deaths_df.copy()

# Fill province first (good practice)
df["Province/State"] = df["Province/State"].fillna("All Provinces")

# Melt (wide → long)
df_long = df.melt(
    id_vars=["Province/State", "Country/Region", "Lat", "Long"],
    var_name="Date",
    value_name="Deaths"
)

# Convert Date to datetime
df_long["Date"] = pd.to_datetime(df_long["Date"], format='%m/%d/%y')

# Aggregate country-level
df_long = df_long.groupby(["Date", "Country/Region"])["Deaths"].sum().reset_index()

# Pivot (final clean format)
df_final = df_long.pivot(index="Date", columns="Country/Region", values="Deaths")

df_final.head()
```

# 7.1 Merge confirmed, deaths, recovered datasets

The datasets were merged using inner joins on 'Date' and 'Country/Region' to create a unified dataset containing confirmed, death, and recovery counts.

```python
# Create long format for confirmed
conf_long = confirmed_df.melt(
    id_vars=["Province/State", "Country/Region", "Lat", "Long"],
    var_name="Date",
    value_name="Confirmed"
)

# Create long format for deaths
death_long = deaths_df.melt(
    id_vars=["Province/State", "Country/Region", "Lat", "Long"],
    var_name="Date",
    value_name="Deaths"
)

# Create long format for recovered
rec_long = recovered_df.melt(
    id_vars=["Province/State", "Country/Region", "Lat", "Long"],
    var_name="Date",
    value_name="Recovered"
)

# Convert Date
for df in [conf_long, death_long, rec_long]:
    df["Date"] = pd.to_datetime(df["Date"], format='%m/%d/%y')

# Aggregate to country level
conf_long = conf_long.groupby(["Date","Country/Region"])["Confirmed"].sum().reset_index()
death_long = death_long.groupby(["Date","Country/Region"])["Deaths"].sum().reset_index()
rec_long = rec_long.groupby(["Date","Country/Region"])["Recovered"].sum().reset_index()

# ✅ Merge all
merged_df = conf_long.merge(death_long, on=["Date","Country/Region"]) \
                     .merge(rec_long, on=["Date","Country/Region"])

merged_df.head()
```

# 7.2 Monthly Sum Analysis

Monthly aggregation helps in understanding broader trends and reduces noise from daily fluctuations.

```python
merged_df["Month"] = merged_df["Date"].dt.to_period("M")

monthly = merged_df.groupby(["Month","Country/Region"])[
    ["Confirmed", "Deaths", "Recovered"]
].sum().reset_index()
```

```python

country = "India"  # change if needed

df_plot = monthly[monthly["Country/Region"] == country]

plt.figure(figsize=(12,6))
plt.plot(df_plot["Month"].astype(str), df_plot["Confirmed"], label="Confirmed")
plt.plot(df_plot["Month"].astype(str), df_plot["Deaths"], label="Deaths")
plt.plot(df_plot["Month"].astype(str), df_plot["Recovered"], label="Recovered")

plt.xticks(rotation=45)
plt.legend()
plt.title(f"Monthly COVID Trends - {country}")
plt.show()
```

# 8.1 Top 3 Countries with Highest Average Death Rate (2020)

*The countries with the highest average death rates indicate higher severity, possibly due to weaker healthcare systems, delayed response, or older population demographics.*

```python
# Filter only 2020 data (FIXED)
df_2020 = merged_df[merged_df["Date"].dt.year == 2020].copy()

# Calculate death rate
df_2020["Death Rate"] = df_2020["Deaths"] / df_2020["Confirmed"]

# Replace inf / NaN
df_2020["Death Rate"] = df_2020["Death Rate"].replace([float('inf'), -float('inf')], 0).fillna(0)

# Average death rate per country
avg_death_rate = df_2020.groupby("Country/Region")["Death Rate"].mean()

# Top 3
top3 = avg_death_rate.sort_values(ascending=False).head(3)

print(top3)
```
